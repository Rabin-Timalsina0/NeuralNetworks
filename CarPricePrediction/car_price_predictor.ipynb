```python
def extract_hp(text):
    match = re.search(r'(\d+\.?\d*)HP', text)
    return float(match.group(1)) if match else None
```

```python
df['horsepower'] = df['engine'].apply(extract_hp)
```

```python
def extract_displacement(text):
    match = re.search(r'(\d+\.\d+)L', text)
    return float(match.group(1)) if match else None

df['engine_size'] = df['engine'].apply(extract_displacement)
```

```python
def extract_cylinders(text):
    match = re.search(r'([V|W](\d+))', text)
    if match:
        return int(match.group(2))
    
    match = re.search(r'Straight (\d+)', text)
    if match:
        return int(match.group(1))
    
    match = re.search(r'(\d+) Cylinder', text)
    if match:
        return int(match.group(1))
    
    return None

df['cylinders'] = df['engine'].apply(extract_cylinders)
```

```python
def categorize_transmission(text):
    if isinstance(text, str):
        if 'a/t' in text.lower() or 'auto' in text.lower() or 'cv' in text.lower() or 'at' in text.lower():
            return 'automatic'
        elif 'm/t' in text.lower() or 'manual' in text.lower() or 'mt' in text.lower():
            return 'manual'
        elif 'f' in text.lower():  
            return 'automatic'
        else:
            return 'automatic'  
    return None
    
def extract_speed(text):
    match = re.search(r'(\d+)-?Speed', text)  # Look for number followed by Speed
    return int(match.group(1)) if match else None

df['speed'] = df['transmission'].apply(extract_speed)

df['transmission'] = df['transmission'].apply(categorize_transmission)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop('price',axis=1))
```

```python
X = scaler.transform(df.drop('price',axis=1))
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,df.price,test_size=0.1)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout, Input
from tensorflow.keras.optimizers import Adam

model = Sequential([
    Input(shape=(72,)),
    Dense(128, activation='relu'),
    BatchNormalization(), 
    Dense(64, activation='relu'),
    Dropout(0.3), 
    Dense(32, activation='relu'),
    Dense(1, activation=None) 
])

model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])  

model.summary()
```

```python
model.fit(X_train, y_train,epochs=30)
```

```python
model.evaluate(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
def predict(brand,year,mileage,fuel_type,transmission,accident,speed,horsepower,engine_size,cylinders):
    arr1 = np.zeros(df.columns[8:-8].shape[0])
    arr1[np.where(df.columns[8:-8]==brand)] = 1 
    arr2 = np.zeros(df.columns[-8:-2].shape[0])
    arr2[np.where(df.columns[-8:-2]==fuel_type)] = 1 
    arr3 = np.zeros(df.columns[-2:].shape[0])
    arr3[np.where(df.columns[-2:]==transmission)] = 1 
    
    features = np.array([year,mileage,accident,horsepower,engine_size,cylinders,speed])
    final_features = np.concatenate((features, arr1, arr2, arr3))
    final_features = final_features.reshape(1, -1)  

    return scaler.transform(final_features)

    
```
